# Digitization/Conversion and Reconstruction of emulsion hit simulations

## Introduction

This procedure follows a SNDSW simulation with emulsion films active.

The EmulsionDetPoints are converted into EdbSegP (FEDRA objects) for reconstruction.

The FairShip2FEDRA.rootrc is used to set the parameters (default values reported here)

They are stored if:

* They are from charged particles;
* The momentum is larger than 100 MeV/c (**FairShip2Fedra.minmomentum**);
* The angle is less than 0.78 rad (**FairShip2Fedra.maxtheta**);

Also, the following real data effects are applied:
* Angular smearing (Gaussian distribution with sigma 3 mrad, **FairShip2Fedra.angres**) is applied when **FairShip2Fedra.dosmearing** is 1  
* Efficiency (Uniform distribution, 85%, **FairShip2Fedra.emuefficiency**) is applied. Alternatively, an angular dependance can be simulated (**FairShip2Fedra.useefficiencymap**);

Finally, the positions and angles are converted in the local brick reference system, **FairShip2Fedra.useLocalBrickCoordinates**
